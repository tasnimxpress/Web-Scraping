# write a function to extract data

```python
data = []
def e(n):
    n_result = r.get(c_url[n])
    n_content = n_result.text
    n_soup = BeautifulSoup(n_content, 'lxml')
    movie_article = n_soup.find('article', attrs={'class': 'main-article'})
    title = movie_article.find('h1').text
    try:
        description = movie_article.find('p', attrs={'class': 'plot'}).text
    except:
        description = 'Not available'
    transcript = movie_article.find('div', attrs={'class': 'full-script'}).text.strip()
    
    return {'title': title,
            'description': description,
            'transcript': transcript}
```

```python
for i in range(len(c_url)):
    extracted_data = e(i)
    data.append(extracted_data)
```
