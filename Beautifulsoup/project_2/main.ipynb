```python
# data dource
url = 'https://www.google.com/finance'
```

```python
# import libraries
import requests as r
from bs4 import BeautifulSoup
```

# specify ticker and exchange

```python
# https://www.google.com/finance/quote/GOOGL:NASDAQ
```

```python
# make price to all in same category
# CAD-USD
def currency_rate(source):
    result = r.get(f'https://www.google.com/finance/quote/{source}-USD')
    content = result.text
    soup = BeautifulSoup(content, 'lxml')

    div = soup.find('div', attrs={'data-source': True})
    rate = float(div['data-last-price'])
    return rate
```

```python
currency_rate('CAD')
```

```python
def get_data(ticker, exchange):
    result = r.get(f'https://www.google.com/finance/quote/{ticker}:{exchange}')
    content = result.text
    soup = BeautifulSoup(content, 'lxml')

    price_div = soup.find('div', attrs={'data-last-price': True})
    currency = price_div['data-currency-code']
    price = float(price_div['data-last-price'])
    
    usd_price = price
    if currency != 'USD':
        usd_price = round(currency_rate(currency) * price, 2)

    return {
        'ticker': ticker,
        'exchange': exchange,
        'currency': currency,
        'price': price,
        'usd_price': usd_price}
```

```python
tick_exchange = {
    'GOOGL':'NASDAQ',
    'META':'NASDAQ',
    'MSFT':'NASDAQ',
    'AMZN':'NASDAQ',
    'CAD':'CVE',
    'SHOP':'TSE'
}
```

```python
data = []
def store_data(ticker, exchange): 
    if not any((d['exchange']==exchange and d['ticker']==ticker) for d in data):
        data.append(get_data(ticker, exchange))
```

```python
for i in tick_exchange:
    store_data(i, tick_exchange[i])
```

```python
data
```

```python
import pandas as pd
df = pd.DataFrame(data)
df.to_csv('stockPrice.csv')
```

```python
# create a function
def get_link(n):
        href = ul[n].find('a', attrs={'href': True})
        page_link = href['href']
        return page_link
```
