```python
root = 'https://books.toscrape.com/'
```

```python
# import libraries
import requests as r
from bs4 import BeautifulSoup
```

```python
result = r.get(root, 'lxml')
result
```

```python
content = result.text
content
```

```python
soup = BeautifulSoup(content)
```

```python
print(soup.prettify)
```

```python
# get the main article
main_article = soup.find_all('article', attrs={'class': 'product_pod'})
len(main_article)
```

```python
# each book page url
page_url = main_article[0].find('a', attrs={'href': True})
page_url['href']
```

```python
# define a function to retrieve all href
def url(n):
    page_url = main_article[n].find('a', attrs={'href': True})
    return page_url['href']
```

```python
print(url(2))
```

```python
# get all the href
url_param = []
for i in range(len(main_article)):
    leftover = url(i)
    url_param.append(leftover)
```

```python
url_param
```

```python
len(url_param)
```

```python
# complete url
c_url = [(root + link) for link in url_param]
len(c_url)
```

```python
# access the new page from complete url
n_result = r.get(c_url[0])
n_result.status_code
```

```python
Ncontent = n_result.text
Ncontent
```

```python
Nsoup = BeautifulSoup(Ncontent, 'lxml')
Nsoup
```

```python
# access the main div
title = Nsoup.find('div', attrs={'class': 'col-sm-6 product_main'}).find('h1').text
title
```

```python
price = Nsoup.find('div', attrs={'class': 'col-sm-6 product_main'}).find('p').text
price
```

```python
# clean the price data
import re # regex
re.sub('[^0-9.]', "", price)
```

```python
# rating
rating = Nsoup.find('p', attrs={'class': 'star-rating'})
rating['class'][1]
```

```python
# UPC
upc = Nsoup.find('table', attrs={'class': 'table table-striped'}).find('td').text
upc
```

```python
# details = Nsoup.find_all('table', attrs={'class': 'table table-striped'})
details = Nsoup.select('table > tr > td')
details[5].text
```

```python
# function for rating
def get_rating(rating):
    return rating['class'][1]
```

```python
# convert to number
def convertToNumber(string):
    number = {
        'One': 1,
        'Two': 2,
        'Three': 3,
        'Four': 4,
        'Five': 5
    }
    return number[string]
```

```python
# clean price data function
def clean_price(value):
    return float(re.sub('[^0-9.]', "", value))

clean_price(price)
```

```python
# clean availability
def stock(num):
    return int(re.sub('[^0-9]', "", num))

stock(details[5].text)
```

```python
# create a function to retreive all books information
def ProductDetails(Nsoup):
    title = Nsoup.find('div', attrs={'class': 'col-sm-6 product_main'}).find('h1').text
    price = Nsoup.find('div', attrs={'class': 'col-sm-6 product_main'}).find('p').text
    rating = Nsoup.find('p', attrs={'class': 'star-rating'})
    details = Nsoup.select('table > tr > td')
    upc = details[0].text
    type = details[1].text
    Price_excl_tax = details[2].text
    Price_incl_tax = details[3].text
    tax = details[4].text
    Availability = details[5].text
    Number_of_reviews = details[6].text

    return {'title': title,
            'price': float(clean_price(price)),
            'rating': convertToNumber(get_rating(rating)),
            'upc': upc,
            'type': type,
            'Price_excl_tax': float(clean_price(Price_excl_tax)),
            'Price_incl_tax': float(clean_price(Price_incl_tax)),
            'tax': float(clean_price(tax)),
            'Availability': stock(Availability),
            'Number_of_reviews': int(Number_of_reviews)}
```

```python
def get_soup(i):
    result = r.get(c_url[i])
    content = result.text
    soup = BeautifulSoup(content, 'lxml')
    return soup
```

```python
booksdata = []
for i in range(len(c_url)):
    soup = get_soup(i)
    data = ProductDetails(soup)
    booksdata.append(data)
```
